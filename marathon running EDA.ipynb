```python
#data cleaning
```

```python
#only USA Races, 50Mi or 50k, 2020
```

```python
#combine 50k/5mi with isin
```

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
#filter USA events
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all the filters to have only USA events ,50Mi or 50k in 2020 in the dataframe
```

```python
df_usa = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df_usa.head(5)
```

```python
df_usa.shape
```

```python
#remove (USA) from event name
```

```python
df_usa['Event name'] = df_usa['Event name'].str.split('(').str.get(0)
```

```python
df_usa.head(5)
```

```python
#clean up athlete age
#calculate age from year of birth
```

```python
df_usa['athlete age'] = 2020 - df_usa['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df_usa['Athlete performance'] = df_usa['Athlete performance'].str.split(' ').str.get(0)
```

```python
df_usa.head()
```

```python
#drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
```

```python
df_usa = df_usa.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df_usa.head()
```

```python
#clean up null values 
```

```python
df_usa.isna().sum()
```

```python
df_usa[df_usa['athlete age'].isna() == 1]
```

```python
df_usa = df_usa.dropna()
```

```python
df_usa.head()
```

```python
df_usa.shape
```

```python
#check for duplicates 
```

```python
df_usa[df_usa.duplicated() == 1]
```

```python
#reset index
```

```python
df_usa.reset_index(drop = True)
```

```python
#fix types
```

```python
df_usa.dtypes
```

```python
df_usa['athlete age'] = df_usa['athlete age'].astype(int) 
```

```python
df_usa['Athlete average speed'] = df_usa['Athlete average speed'].astype(float) 
```

```python
df_usa.dtypes
```

```python
#rename columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete age                    int32
```

```python
df_usa = df_usa.rename(columns = {'Year of event':'year',
                                  'Event dates':'race_day',
                                  'Event name':'race_name',
                                  'Event distance/length':'race_length',
                                  'Event number of finishers':'race_number_of_finishers',
                                  'Athlete performance':'athlete_performance',
                                  'Athlete gender':'athlete_gender',
                                  'Athlete average speed':'athlete_average_speed',
                                  'Athlete ID':'athlete_id',
                                  'athlete age':'athlete_age'
                                  })
```

```python
df_usa.head()
```

```python
#reorder columns
```

```python
df_usa = df_usa[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed','year']]
```

```python
#difference in speed for 50k, 50mi male to female
```

```python
df_usa.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#10 best age groups in the 50mi race (with at least 20 athletes in the group)
```

```python
df_usa.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head(10)
```

```python
#20 worst age groups in the 50mi race (with at least 10 athletes in the group)
```

```python
df_usa.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9').head(20)
```

```python
#check if athletes are slower in different seasons

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2
```

```python
df_usa['race_month'] = df_usa['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df_usa['race_season'] = df_usa['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring')
```

```python
#athletes average speed in different seasons for 50mi and 50km
```

```python
df_usa.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#athletes average speed in different seasons for 50mi only
```

```python
df_usa.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#athletes average speed in different seasons for 50km only
```

```python
df_usa.query('race_length == "50km"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
